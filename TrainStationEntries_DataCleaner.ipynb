## Comparablity with bus and tram 'boarding and alighting' dataset
This dataset does not provide boarding or departure information, but does record myki transations to enter or leave a railway station.  The 'Normal Weekday' result records the whole day.

For comparison with bus and tram results use the Weekeday time band breakdown for 2011 - 2012 
Time bands are not specified on Data Vic Gov Au site but are on the PTV Research website https://www.ptv.vic.gov.au/about-ptv/ptv-data-and-reports/research-and-statistics/

Pre AM Peak - first service to 6:59am
AM Peak - 7:00am to 9:29am
Interpeak - 9:30am to 2:59pm
PM Peak - 3:00pm to 6:59pm
PM Late - 7:00pm to last service

Use SUM('AM Peak', 'Interpeak', 'PM Peak') for a comparable slice of the day to the buses and tram boading data.
