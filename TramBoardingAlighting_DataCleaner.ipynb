# Tram Boarding and Alighting Data Cleaner

Tram Boardings and Alightings at Tram Stops
https://www.data.vic.gov.au/data/dataset/tram-boardings-and-alightings-at-tram-stops-2015
Download manually, work on local copy.

### Goal: To identify busiest (most boardings and alightings) tram stops on the network
#### Secondary goal: To create a common measure for all public transport modes to indentify busiest Public Transport stops
Tram stops are uniquely identifed using Metlink_Stop_ID. 
(A tram route with a stop on either side of the road will have 2 Metlink Stop IDs but only one Tram Tracker ID)
The same tram stop can be used by multiple trams

```python
import pandas as pd
```

```python
rawtram = './raw/Tram Boardings and Alightings 2011 - data.XLS'
```

```python
df = pd.read_excel(rawtram,sheetname='Data', header=0,converters={'Route_Number':str,'Tram_Tracker_ID':str, 'Metlink_Stop_ID':str, 'VicgridX':str, 'VicgridY':str})
df
```

```python
bystop = df.groupby('Metlink_Stop_ID').sum()
bystop['DailyUse'] = bystop.Boardings.fillna(0) + bystop.Alightings.fillna(0)

bystop.to_csv('./clean/TramStopTraffic.csv')
bystop
```

## Data quality note: 
The data definitions worksheet notes there should be a column "Day_Type", "Day type either weekday , Saturday or Sunday".
This column appears to be missing from the data set.

However, Most of the tram boarding / alighting data appears to be grouped into collections of three rows (Route then direction).   Data that is not in a group of three is a single row.  

I have assumed that the first of three rows is "Weekday".  
The second and third (when they exist) are "Saturday" and "Sunday" respectively.  This order matches the order in the meta data description and appars to be a reasonable assumption to make.

## Coordinate Reference System Note:
Columns H and I provide X and Y coordinates in VicGrid Coordinate Reference System. 
I believe VicGrid to be a subset of GDA2020 EPSG:7899.
As of August 2017 I've not been able to find support for this coordinate reference system under QGIS.

A point layer of Melbourne Tram stops in CRS=GDA94 format is available from
https://www.data.vic.gov.au/data/dataset/public-transport-a-collection-of-ptv-datasets
